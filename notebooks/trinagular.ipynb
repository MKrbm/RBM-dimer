```python
from netket.graph import Triangular
import matplotlib.pyplot as plt
import numpy as np

g = Triangular(extent=[3, 3])
```

```python
from netket.graph._lattice_edge_logic import create_padded_sites
```

```python
pos, labels = create_padded_sites(g.basis_vectors, g.extent, g.site_offsets, g.pbc, 1)

```

```python
# g._basis_coord_to_site
g.basis_vectors
```

```python
# Extracting x and y coordinates
x = pos[:, 0]
y = pos[:, 1]

# Creating the plot
plt.figure(figsize=(8, 8))
scatter = plt.scatter(x, y, c=labels, cmap='viridis')  # Color code by label
plt.colorbar(scatter, label='Site Labels')  # Add colorbar with label

# Annotating each point with its label
for i, label in enumerate(labels):
    plt.annotate(label, (x[i], y[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.title('Triangular Lattice with Padded Sites')
plt.xlabel('X coordinate')
plt.ylabel('Y coordinate')
plt.grid(False)
plt.axis('equal')
plt.show()
```

```python
positions = g.positions

# Extracting x and y coordinates
x = positions[:, 0]
y = positions[:, 1]


# Assuming g.positions and g.edges(0) are already defined
positions = g.positions
edges = g.edges(0)

# Extracting x and y coordinates
x = positions[:, 0]
y = positions[:, 1]

# Creating the plot
plt.figure(figsize=(6, 6))
plt.scatter(x, y, color='blue')  # Plot points

# Plotting edges with length condition
for edge in edges:
    start, end = edge
    # Calculate the Euclidean distance between points
    distance = np.linalg.norm(positions[start] - positions[end])
    if distance <= 1:
        plt.plot([positions[start][0], positions[end][0]], [positions[start][1], positions[end][1]], 'k-')  # 'k-' for black line

plt.title('Triangular Lattice Points with Edges')
plt.xlabel('X coordinate')
plt.ylabel('Y coordinate')
plt.grid(False)
plt.axis('equal')
plt.show()

```
